```python
from IPython.display import HTML
import base64
import io

def save_plot(plt, description):
    """Сохраняет текущий график в HTML без очистки"""
    buf = io.BytesIO()
    plt.savefig(buf, format='png', bbox_inches='tight')
    buf.seek(0)
    img_base64 = base64.b64encode(buf.read()).decode('utf-8')

    html = f"""
    <div style="margin: 20px; padding: 15px; border: 1px solid #eee; border-radius: 5px;">
        <h3>{description}</h3>
        <img src='data:image/png;base64,{img_base64}' width="80%">
    </div>
    """
    plt.close()  # Закрываем только после сохранения
    return html
```

```python
dframe = pd.read_excel(data)
dframe = dframe.drop(['количество карточек', 'pog', 'posl', 'ran','s_dtp',], axis=1)
dframe = dframe.drop(['Дорога категория'], axis=1)
dframe = dframe.drop(['Тип детского удерживающего устройства'], axis=1)
dframe = dframe.drop(['Степень опьянения2'], axis=1)
# Преобразуем дату в datetime
dframe['Дата'] = pd.to_datetime(dframe['Дата'], format='%d.%m.%Y')

month_names = {
    1: 'Январь', 2: 'Февраль', 3: 'Март', 4: 'Апрель',
    5: 'Май', 6: 'Июнь', 7: 'Июль', 8: 'Август',
    9: 'Сентябрь', 10: 'Октябрь', 11: 'Ноябрь', 12: 'Декабрь'
}
# Извлекаем месяц (1-12)
dframe['Месяц'] = dframe['Дата'].dt.month.map(month_names)

# Преобразуем время в datetime и извлекаем час
dframe['Время_datetime'] = pd.to_datetime(dframe['Время'], format='%H:%M').dt.time
dframe['Час'] = pd.to_datetime(dframe['Время'], format='%H:%M').dt.hour

# Определяем периоды дня
def get_time_period(hour):
    if 5 <= hour < 12:
        return 'Утро'
    elif 12 <= hour < 17:
        return 'День'
    elif 17 <= hour < 22:
        return 'Вечер'
    else:
        return 'Ночь'

dframe['Период_дня'] = dframe['Час'].apply(get_time_period)
# dframe = dframe.drop(['Дата_datetime'], axis=1)
dframe = dframe.drop(['Время_datetime'], axis=1)
```

```python
# Правильный порядок месяцев
month_order = ['Январь', 'Февраль', 'Март', 'Апрель', 'Май', 'Июнь',
               'Июль', 'Август', 'Сентябрь', 'Октябрь', 'Ноябрь', 'Декабрь']

# Создаем DataFrame с правильным порядком месяцев
monthly_accidents = (dframe.groupby('Месяц')['Номер ДТП']
                     .count()
                     .reindex(month_order)
                     .dropna())  # Удаляем месяцы, которых нет в данных

# Строим график
plt.figure(figsize=(12, 6))
monthly_accidents.plot(kind='bar', color='steelblue', title='Количество ДТП по месяцам')

# Настройки графика
plt.xlabel('Месяц', fontsize=12)
plt.ylabel('Количество ДТП', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plot1_html = save_plot(plt, "Рис. 1: Распределение ДТП по месяцам")
```

```python
time_period_accidents = dframe.groupby('Период_дня')['Номер ДТП'].count()
time_order = ['Утро', 'День', 'Вечер', 'Ночь']
time_period_accidents = time_period_accidents.reindex(time_order)

time_period_accidents.plot(kind='pie', autopct='%1.1f%%',
                         title='ДТП по времени суток')

plot2_html = save_plot(plt, "Рис. 2: ДТП по времени суток")
```

```python
# Преобразуем столбцы в категориальные с правильным порядком
# Группируем данные с сохранением порядка
month_time = (dframe.groupby(['Месяц', 'Период_дня'], observed=True)['Номер ДТП']
              .count()
              .unstack()
              .reindex(month_order)
              [time_order])  # Упорядочиваем столбцы по времени суток

# Строим график
plt.figure(figsize=(15, 10))
month_time.plot(kind='area', stacked=True,
                title='ДТП по месяцам и времени суток',
                alpha=0.8,  # Прозрачность для лучшей видимости
                cmap='viridis')  # Цветовая схема

# Настройки графика
plt.xlabel('Месяц', fontsize=12)
plt.ylabel('Количество ДТП', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Время суток', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()

plot3_html = save_plot(plt, "Рис. 3: Распределение ДТП по месяццам и времени суток")
```

```python
import folium
from folium.plugins import MarkerCluster, Search
from folium import FeatureGroup

# 1. Удаляем дубликаты ДТП (если 'Номер ДТП' уникален)
unique_dtp = dframe.drop_duplicates(subset=['Номер ДТП'])

# 2. Группируем по адресу и возрасту, считаем ДТП и берем координаты
dangerous_locations = unique_dtp.groupby(['Адрес', 'Промежуток возрастов']).agg({
    'Номер ДТП': 'count',
    'Долгота': 'first',
    'Широта': 'first'
}).reset_index()

# 3. Удаляем записи без координат
dangerous_locations = dangerous_locations.dropna(subset=['Широта', 'Долгота'])

# Топ-10 самых опасных адресов
top_dangerous = dangerous_locations.groupby('Адрес')['Номер ДТП'].sum().nlargest(10)

# 4. Создаем карту
map_center = [dangerous_locations['Широта'].mean(), dangerous_locations['Долгота'].mean()]
m = folium.Map(location=map_center, zoom_start=10)

# 5. Создаем ОТДЕЛЬНЫЕ СЛОИ ДЛЯ КАЖДОЙ ВОЗРАСТНОЙ ГРУППЫ
age_groups = dangerous_locations['Промежуток возрастов'].unique()
layer_dict = {}  # Словарь для хранения слоев

for age in age_groups:
    layer_dict[age] = FeatureGroup(name=f"Возраст: {age}", show=True)
    m.add_child(layer_dict[age])

# 6. Создаем кластеры внутри каждого слоя
cluster_dict = {}
for age in age_groups:
    cluster_dict[age] = MarkerCluster().add_to(layer_dict[age])

# 7. Добавляем маркеры в соответствующие кластеры
for _, row in dangerous_locations.iterrows():
    age = row['Промежуток возрастов']
    popup_text = f"""
    <b>Адрес:</b> {row['Адрес']}<br>
    <b>Возраст:</b> {age}<br>
    <b>Кол-во ДТП:</b> {row['Номер ДТП']}
    """

    folium.CircleMarker(
        location=[row['Широта'], row['Долгота']],
        radius= 3 + row['Номер ДТП'] * 0.5,
        popup=popup_text,
        color='red' if age in young_groups else 'orange',
        fill=True,
        fill_opacity=0.7,
        tooltip=f"{row['Адрес']} | {age}"
    ).add_to(cluster_dict[age])


# 8.  Легенда и управление слоями
legend_html = '''
     <div style="position: fixed; bottom: 50px; left: 50px; width: 200px; z-index:9999;
                 background:white; padding: 10px; border: 2px solid grey;">
     <b>Легенда:</b><br>
     &bull; <i style="color:red">●</i> Молодые группы<br>
     &bull; <i style="color:orange">●</i> Другие возрасты
     </div>
'''
m.get_root().html.add_child(folium.Element(legend_html))
folium.LayerControl().add_to(m)


# 9. Сохраняем карту
m.save('ДТП_с_группировкой_по_возрасту_и_адресу.html')
print(top_dangerous)
m
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer

df_b = df.copy()

# 1. Подготовка целевой переменной: бинарный показатель "высокий риск"
# Допустим, высокий риск = ДТП с тяжелыми последствиями или смертельным исходом
df_b['высокий_риск'] = df_b['ТС Степень тяжести последствий'].isin([
'Раненый, находящийся (находившийся) на стационарном лечении',
'Скончался в течение 1 суток',
'Скончался в течение 12 суток',
'Скончался в течение 17 суток',
'Скончался в течение 23 суток',
'Скончался в течение 3 суток',
'Скончался в течение 5 суток',
'Скончался в течение 8 суток',
'Скончался на месте ДТП до приезда скорой медицинской помощи',
'Скончался на месте ДТП по прибытию скорой медицинской помощи, но до транспортировки в мед. организацию'
]).astype(int)

df_young = df_b[df_b['Промежуток возрастов'].isin(young_groups)].copy()

# 3. Отбор признаков
features = [
'Адрес', 'Дорога наименование', 'Дорога значение',
'Факторы, оказывающие влияние на режим движения',
'Улица категория', 'Населенный пункт', 'НДУ',
'Объекты УДС вблизи места ДТП', 'Освещение', 'Состояние проезжей части',
'Объекты УДС на месте ДТП', 'погода', 'Улица',
'Год выпуска', 'модель ТС', 'Марка', 'Расположение руля, тип привода',
'Технические неисправности', 'Тип ТС','ТС Степень опьянения',
'ТС Непосредственные нарушения ПДД', 'ТС Пол', 'Использовался ли ремень',
'ТС Сопутствующие нарушения ПДД', 'Тип детского удерживающего устройства',
'Водительский стаж', 'Месяц', 'Час', 'Период_дня'
]

X = df_young[features]
y = df_young['высокий_риск']

# 4. Разделение данных
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y)

# 5. Создание pipeline для обработки данных

numeric_features = [var for var in features if X[var].dtype!='O']
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])

categorical_features = [var for var in features if X[var].dtype=='O']
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])

# 6. Обучение моделей
models = {
    "Random Forest": RandomForestClassifier(
        n_estimators=200,
        class_weight='balanced',
        random_state=42),

    "Gradient Boosting": GradientBoostingClassifier(
        n_estimators=150,
        learning_rate=0.1,
        max_depth=5,
        random_state=42)
}

results = {}
for name, model in models.items():
    clf = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('classifier', model)])

    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    y_proba = clf.predict_proba(X_test)[:, 1]

    results[name] = {
        'model': clf,
        'report': classification_report(y_test, y_pred),
        'roc_auc': roc_auc_score(y_test, y_proba),
        'f1': f1_score(y_test, y_pred)
        }

# 7. Вывод результатов (исправленная версия)
report_parts = []
for model_name, result in results.items():
    print(f"\n{model_name} Performance:")
    print("="*50)
    print(f"ROC-AUC: {result['roc_auc']:.3f}")
    print(f"F1 Score: {result['f1']:.3f}")
    print("\nClassification Report:")
    print(result['report'])
    report = result['report']
    matrix = confusion_matrix(y_test, y_pred)
    print(confusion_matrix(y_test, y_pred))

    # Используем pipeline (clf) вместо model для вычисления точности
    clf = result['model']

    # Точность моделирования с использованием тестовых данных
    # Вычисление accuracy (для классификации это более подходящая метрика, чем R²)
    accuracy = clf.score(X_test, y_test)

    # Вычисление средней абсолютной ошибки между исходными и предсказанными значениями
    MAE_test = mean_absolute_error(y_test, y_pred)
    # Вычисление средней квадратической ошибки между исходными и предсказанными значениями
    MSE_test = mean_squared_error(y_test, y_pred)
    # Корень из средней квадратической ошибки между исходными и предсказанными значениями
    RMSE_test = np.sqrt(MSE_test)

    print('Точность тестового набора:')
    print("Accuracy на тестовых данных =", accuracy)
    print("MAE на тестовых данных =", MAE_test)
    print("MSE на тестовых данных =", MSE_test)
    print("RMSE на тестовых данных =", RMSE_test)

    report_html = save_classification_results(
        report,
        matrix,
        accuracy,  # Используем accuracy вместо R²
        MAE_test,
        MSE_test,
        RMSE_test,
        f"{model_name} (AUC: {result['roc_auc']:.3f}, F1: {result['f1']:.3f})"
    )
    report_parts.append(report_html)

# 8. Визуализация важности признаков (для Random Forest)
rf_model = results['Random Forest']['model']
feature_names = (rf_model.named_steps['preprocessor']
                 .named_transformers_['cat']
                 .named_steps['onehot']
                 .get_feature_names_out(categorical_features))

all_features = numeric_features + list(feature_names)
importances = rf_model.named_steps['classifier'].feature_importances_

report3 = report_parts
```
